```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds=xr.open_dataset("CABLE_AUX-dev/offline/CABLE_UNSW_GSWP3_gridinfo_0.5x0.5.nc")
```

`iveg` ranges from 1 to 17.

LDT data comes with fractions for each PFT in each gridcell. Here we want only the dominant PFT. So we take the `argmax` along the `sfctypes` and add 1 to start from 1 as the PFTs.

```python
ldt=xr.open_dataset("/g/data/w35/ccc561/temp/lis_input.d01.nc")
```

```python
lat = ldt["lat"][:,0]
lon = ldt["lon"][0,:]
```

```python
sfc=ldt["LANDCOVER"].argmax(dim="sfctypes")+1
sfc=sfc.assign_coords({"north_south":lat})
sfc=sfc.assign_coords({"east_west":lon})
sfc=sfc.rename({"north_south":"latitude","east_west":"longitude"})
```

## PFT 1

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(10,10))
opts={'cmap':"tab20",'vmin':1,'vmax':17}
ds["iveg"].sel(longitude=slice(0,180)).where(ds["iveg"]==1).plot(ax=ax1,**opts)
sfc.where((sfc==1)|(sfc==5)).plot(ax=ax2,**opts)
ax1.set_title("CABLE PFT 1")
ax2.set_title("LIS PFT 1+5")
```

## Veg types at the same spatial location

```python
cable = ds["iveg"].sel(longitude=slice(0,180))
for iveg in range(1,18):
    unilist,unicount=np.unique(sfc.where(cable==iveg),return_counts=True)
    unilist=unilist[~np.isnan(unilist)].astype(np.int64)  # Remove NaN values
    unicount=unicount[0:unilist.size]
    sortind=unicount.argsort()  # Sort in descending order
    unilist=unilist[sortind[::-1]]
    unicount=unicount[sortind[::-1]]
    print(f"Veg type in CABLE: {iveg}")
    print(f"(LIS type, # points) {list(zip(unilist,unicount))}")
```

## Remap LIS PFTs

```python
mapveg={1:(1,5),
        2:(2,),
        3:(3,),
        4:(4,),
        5:(7,),
        6:(8,9,10),
        7:(),
        8:(18,19),
        9:(12,),
        10:(),
        11:(11,),
        12:(),
        13:(),
        14:(16,),
        15:(13,),
        16:(17,),
        17:(15,)
       }
```

```python
sfc1=ldt["LANDCOVER"]
sfc1=sfc1.assign_coords({"north_south":lat})
sfc1=sfc1.assign_coords({"east_west":lon})
sfc1=sfc1.rename({"north_south":"latitude","east_west":"longitude"})
```

```python
new_map = sfc1.copy()
new_map[:,:,:] = 0.

for iveg,comb in mapveg.items():
    if len(comb)!=0:
        for ii in comb:
            new_map[iveg-1,...]=new_map[iveg-1,...]+sfc1[ii-1,...]
```

```python
ldtveg_map=new_map.argmax(dim="sfctypes")+1
```

```python
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(10,10))
opts={'cmap':"tab20",'vmin':1,'vmax':17}
ds["iveg"].sel(longitude=slice(0,180)).plot(ax=ax1,**opts)
ldtveg_map.plot(ax=ax2,**opts)
ax1.set_title("CABLE PFTs")
ax2.set_title("Remapped LIS PFTs")
```
